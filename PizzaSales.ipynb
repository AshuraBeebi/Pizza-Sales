```python
import matplotlib.pyplot as plt
```

```python
y=df.loc[0:80,'Unit_Price']
plt.plot(y)
plt.show()
```

```python
gen=df['pizza_category'].value_counts()
print(type(gen))
print(gen)
x=gen.values
lb=gen.index
plt.figure(figsize=(8,6),dpi=100)
plt.pie(x,labels=lb)
plt.legend(lb)
plt.title("Category")
plt.show()
```

```python

smoking_stroke=df[df['total_price']>30]['pizza_size']
smoking_series=smoking_stroke.value_counts()

x=smoking_series.index
y=smoking_series.values
plt.figure(figsize=(8,6),dpi=100)
plt.title("Distribution of Smoking Status of persons with Stroke")
plt.xlabel("Smoking Status")
plt.ylabel("Counts ")


plt.bar(x,y,color="red")
plt.show()

```

```python
plt.figure(figsize=(8,6),dpi=100)
x=df['Unit_Price']
plt.hist(x,color='green')
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Counts")
# plt.xticks([0,20,40,60,80])
plt.show()
```

```python
gen=df['pizza_name'].value_counts()
print(type(gen))
print(gen)
x=gen.values
lb=gen.index
plt.figure(figsize=(8,6),dpi=100)
plt.pie(x,labels=lb)
plt.legend(lb)
plt.title("Name")
plt.show()
```

```python
#This groups the entire DataFrame df by pizza name.
popular_pizzas = df.groupby('pizza_name')['quantity'].sum().head(10)
popular_pizzas.plot(kind='bar')
plt.title("Top 10 Best-Selling Pizzas")
plt.ylabel("Total Quantity Sold")
```

```python
#Gives the total revenue per category.
#sort-values=ascending order
category_revenue = df.groupby('pizza_category')['total_price'].sum().sort_values()
category_revenue.plot(kind='barh', color='skyblue')
plt.title("Revenue by Pizza Category")
plt.xlabel("Total Revenue")
```

```python
import matplotlib.pyplot as plt

# Count frequency of each unit price
unit_price_counts = df['Unit_Price'].value_counts().sort_index()

plt.figure(figsize=(8,6), dpi=100)
plt.scatter(unit_price_counts.index, unit_price_counts.values, color='purple')
plt.title("Distribution of Unit Price")
plt.xlabel("Unit Price")
plt.ylabel("Counts")
plt.grid(True)
plt.show()
```

```python
# Filter for Medium size pizzas
pizza_m = df[df['pizza_size'] == 'M']

# Count categories
category_counts = pizza_m['pizza_category'].value_counts()
x = category_counts.values
labels = category_counts.index

# Plot pie chart
plt.figure(figsize=(8, 6), dpi=100)
plt.pie(x, labels=labels)
plt.title("Pizza Category Distribution for Medium Sized Pizzas")
plt.legend(x, loc="upper left")
plt.show()

```
